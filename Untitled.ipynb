```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import calendar
from scipy.stats import zscore

# Load your data and parse date
df = pd.read_csv("data/train_data.csv", skiprows=20000, nrows=20000, header=None)
df['date'] = pd.to_datetime(df[1], errors='coerce')

# Extract month and day
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

# Filter defaulters (assuming label in last column)
label_col = df.columns[-1]
df_defaulters = df[df[label_col] == 1]

# Rename feature column for clarity
feature_col = 2
df_defaulters = df_defaulters.rename(columns={feature_col: 'feature'})

months = sorted(df_defaulters['month'].unique())
month_names = [calendar.month_name[m] for m in months]
days = sorted(df_defaulters['day'].unique())

def detect_outliers_iqr(data):
    """Detect multiple outliers using IQR method."""
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data < lower_bound) | (data > upper_bound)]

def detect_outliers_zscore(data, threshold=2.5):
    """Detect multiple outliers using Z-score method."""
    if len(data) < 2:
        # zscore needs at least 2 data points
        return pd.Series(dtype=data.dtype)
    zs = zscore(data)
    return data[(np.abs(zs) > threshold)]

# Choose detection method here:
outlier_detector = detect_outliers_zscore  # or detect_outliers_iqr

plt.figure(figsize=(22, 10))
ax = sns.boxplot(
    x='day',
    y='feature',
    hue='month',
    data=df_defaulters,
    showfliers=False,
    palette='tab20',
    dodge=True
)

width = 0.8
num_months = len(months)
width_per_month = width / num_months

all_outliers = []

for i, day in enumerate(days):
    for j, month in enumerate(months):
        group = df_defaulters[(df_defaulters['day'] == day) & (df_defaulters['month'] == month)]['feature']
        if group.empty:
            continue
        outliers = outlier_detector(group)
        if not outliers.empty:
            x = i - width/2 + width_per_month/2 + j * width_per_month
            jitter = np.random.uniform(-width_per_month/4, width_per_month/4, size=len(outliers))
            ax.scatter(x + jitter, outliers, color='red', s=30, edgecolor='black', alpha=0.7)
            all_outliers.append((month, day, outliers.values))

# Print summary of detected outliers:
if all_outliers:
    print(f"Detected outliers in total {len(all_outliers)} groups.")
    for month, day, values in all_outliers[:5]:  # print first 5 groups only
        print(f"Month: {calendar.month_name[month]}, Day: {day}, Outliers count: {len(values)}")
        print(f"Sample outlier values: {values[:5]}")
else:
    print("No outliers detected.")

plt.title(f'Boxplot of Feature {feature_col} by Day and Month for Defaulters\n(with multiple outliers detected)', fontsize=16)
plt.xlabel('Day of Month', fontsize=14)
plt.ylabel(f'Feature {feature_col}', fontsize=14)
plt.xticks(rotation=45)

# Expand y-axis limits a bit so outliers aren’t cut off
ylim = ax.get_ylim()
ax.set_ylim(ylim[0] - 0.1 * abs(ylim[0]), ylim[1] + 0.1 * abs(ylim[1]))

# Legend
handles, labels = ax.get_legend_handles_labels()
from collections import OrderedDict
unique_labels = OrderedDict(zip(labels, handles))
plt.legend(unique_labels.values(), unique_labels.keys(), title='Month', bbox_to_anchor=(1.05, 1),
           loc='upper left', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("data/train_data.csv", skiprows=20000, nrows=20000, header=None)
df['date'] = pd.to_datetime(df[1], errors='coerce')

# Drop rows where date is NaT
df = df.dropna(subset=['date'])

# Extract month and day
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

# Filter defaulters
label_col = df.columns[-1]
df_defaulters = df[df[label_col] == 1]

# Plot histogram
plt.figure(figsize=(16, 6))
sns.histplot(
    data=df_defaulters,
    x='month',
    bins=12,
    multiple='stack',  # Shows stacked bars per month by default
    hue='day',         # Colors indicate day of month
    palette='viridis',
    shrink=0.8
)

plt.xlabel("Month")
plt.ylabel("Number of Defaulters")
plt.title("Histogram of Credit Card Defaulters by Month and Day")
plt.xticks(range(1, 13))
plt.legend(title="Day of Month", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Step 1: Keep only numeric columns for correlation
features_numeric = features_encoded.select_dtypes(include=['int64', 'float64'])

# Step 2: Compute the correlation matrix
correlation_matrix = features_numeric.corr()

# Step 3: Unstack the matrix to get pairwise correlations
correlation_pairs = correlation_matrix.unstack()

# Step 4: Remove self-correlations
correlation_pairs = correlation_pairs[correlation_pairs.index.get_level_values(0) != correlation_pairs.index.get_level_values(1)]

# Step 5: Drop duplicate pairs (A-B is same as B-A)
correlation_pairs = correlation_pairs.groupby(lambda x: tuple(sorted(x))).mean()

# Step 6: Sort by absolute correlation value
sorted_correlation = correlation_pairs.sort_values(ascending=False)

# Step 7: Filter for high and low correlations
high_corr = sorted_correlation[abs(sorted_correlation) > 0.8]
low_corr = sorted_correlation[abs(sorted_correlation) < 0.2]

# Display
print("🔴 Highly Correlated Pairs (|corr| > 0.8):\n")
print(high_corr)

print("\n🟢 Weakly Correlated Pairs (|corr| < 0.2):\n")
print(low_corr)
```

```python
from collections import defaultdict

def group_columns_by_dtype(df: pd.DataFrame) -> dict:
    dtype_groups = defaultdict(list)
    for col in df.columns:
        dtype_str = str(df[col].dtype)
        dtype_groups[dtype_str].append(col)
    return dict(dtype_groups)
```

```python
processed_chunk = sample_df
grouped_dtypes = group_columns_by_dtype(processed_chunk)

for dtype, columns in grouped_dtypes.items():
    print(f"{dtype}: {columns}")
```

```python
def flatten_sample_df(df: pd.DataFrame, flatten_cols: list, id_col: str = 'customer_ID') -> pd.DataFrame:
    agg_funcs = ['mean', 'max', 'min', 'std']
    flattened_df = df.groupby(id_col)[flatten_cols].agg(agg_funcs)
    flattened_df.columns = [f"{col}_{agg}" for col, agg in flattened_df.columns]
    flattened_df = flattened_df.reset_index()
    return flattened_df

flattened_output = flatten_sample_df(sample_df, flatten_cols=flatten_cols, id_col='customer_ID')
```
